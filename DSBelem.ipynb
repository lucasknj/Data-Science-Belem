### Verificando correlação

```python
plt.figure(figsize=(22, 16))
correlacao = features.corr()
sns.heatmap(correlacao,linewidths=.5, annot=True)
```

##### Considerações
 - Através do heatmap pode se observar forte correlação entre:
     * 'radius_mean','perimeter_mean','area_mean','radius_worst','perimeter_worst', 'area_worst' ----> 'area_mean'
     * 'compactness_mean', 'concavity_mean','concave points_mean' ----> 'concave points_mean'
     * 'radius_se', 'perimeter_se', 'area_se'---> 'area_se'
     * 'compactness_worst', 'concavity_worst', 'concave points_worst' ---> 'concave points_worst'
     * 'texture_mean','texture_worst' ---> 'texture_mean'
     
 - Depois da primeira análise do heatmap, optei por remover as variáveis com alta correlação. Após a primeira redução, optei por retirar a variável cocave points_worst também.
     * 'concave points_mean','concave points_worst' ---> 'concave points_mean'

```python
#Retirar as variáveis que tenham correlação
features = features.drop(['radius_mean','perimeter_mean','radius_worst','perimeter_worst', 'area_worst',
                         'compactness_mean', 'concavity_mean','radius_se', 'perimeter_se',
                         'compactness_worst', 'concavity_worst', 'concave points_worst','texture_worst'],axis=1)

#Transformar o df features em uma lista (pra ser utilizado no index do df results)
f = list(features)
```

```python
#Select features according to the k highest scores. http://scikit-learn.org/stable/modules/feature_selection.html
from sklearn.feature_selection import SelectKBest,chi2
selector = SelectKBest(k='all').fit(features,target)
print (selector.scores_)

results = pd.DataFrame(selector.scores_, index=f)

results.columns = ['Importancia']
results.sort_values(['Importancia'],ascending=False)
```

```python
plt.figure(figsize=(16, 12))
correlacao = features.corr()
sns.heatmap(correlacao,linewidths=.5, annot=True)
```

#### Split Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

#### Criando modelo XGBoost

```python
#!pip install xgboost
```

```python
#https://machinelearningmastery.com/develop-first-xgboost-model-python-scikit-learn/
#from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)

pred = model.predict(X_test)
acc = accuracy_score(y_test,pred)
acc

#0.9574468085106383 split 0.2
#0.9432624113475178 split 0.3
```

#### Número de features a ser utilizadas

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFECV.html
from sklearn.feature_selection import RFECV

selector = RFECV(XGBClassifier(),step=1, cv=5)
selector = selector.fit(X_train, y_train)

print("Quantidade features",selector.n_features_)
print(results[selector.support_])
print("Lista de Scores:",selector.grid_scores_)
print("High Score",selector.grid_scores_.max())

features_selected = results[selector.support_].index

#0.9625898530109056 split 0.2
#0.9603263403263403 split 0.3
```

```python
sns.lineplot(x=range(1, len(selector.grid_scores_)+1), y=selector.grid_scores_)
plt.xlabel("Número de features selecionadas")
plt.ylabel("Acurácia do modelo")
plt.show()
```

```python
features_selected
```

#### Split Data

```python
features_selecionadas = features[features_selected]

X_train, X_test, y_train, y_test = train_test_split(features_selecionadas, target, test_size=0.3, random_state=42)
```

#### Otimização

```python
#https://www.analyticsvidhya.com/blog/2016/03/complete-guide-parameter-tuning-xgboost-with-codes-python/
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score

parameters = {
    "learning_rate": [0.2], #[0.15, 0.2, 0.3]
    "min_samples_split": [2], #[2,3,4]
    "min_samples_leaf": [10,20], #[20,30,40]
    "max_depth":[3], #[3,5,8]
    "subsample":[0.5], #[0.2,0.5,0.7]
    "n_estimators":[80] #[60,70,80,90]
    }

scor = make_scorer(f1_score,average="micro")
clf = GridSearchCV(XGBClassifier(), parameters, cv=10, scoring = scor)
#{'learning_rate': 0.2, 'max_depth': 3, 'min_samples_leaf': 10, 'min_samples_split': 2, 'n_estimators': 80, 'subsample': 0.5}

clf.fit(X_train, y_train)
print(clf.best_params_)
```

```python
best_clf = clf.best_estimator_

pred = best_clf.predict(X_test)

print("f1-score for test set: {:f}".format(f1_score(y_test,pred,average="micro")))

#0.964539
```
